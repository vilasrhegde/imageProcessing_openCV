# Resizing image using pillow lib

*Some ML libraries require resized images only for processing all the images*



```python
from PIL import Image
```

```python
img = Image.open('/content/dog1.jpg')
img_resz = img.resize((200,200))
```

```python
img_resz.save('dog1_resized.jpg')
```

```python
imgRes = mpimg.imread('/content/dog1_resized.jpg')
imgRes_plot = plt.imshow(imgRes)
plt.show()
```

# Converting RGB to Greyscale using openCV

```python
import cv2
```

```python
img = cv2.imread('/content/dog1.jpg')
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
# coloab won't allow this function
## cv2.imshow() 
```

```python
#saving the grayscale Image
cv2.imwrite('dog1_grayscale.jpg',grayscale_img)
```
